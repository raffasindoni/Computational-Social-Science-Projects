## 2. Data Preprocessing and Cleaning

```python
# drop datetime info
# -----------
chicago_inspections_2011_to_2013 = chicago_inspections_2011_to_2013.dropna().drop(['Inspection_Date',
                                                                                   'minDate',
                                                                                   'maxDate',
                                                                                   'Inspection_MonthYear'],
                                                                                  axis = 1)
```

```python
# process target 
# -----------
y = chicago_inspections_2011_to_2013['Results']

# decide if you want to binarize the outcome variable 
# -----------
# comment out the following lines of code if you don't want to binarize the target variable
y = y.replace({'Pass w/ Conditions': 'Pass'})
lb_style = LabelBinarizer()
y = lb_style.fit_transform(y)

# recode 0s and 1s so 1s are "Fail"
y = np.where(y == 1, 0 ,1)

# Ensure y is 1D
y = y.ravel()

# process features
# -----------

# create feature dataset
X = chicago_inspections_2011_to_2013.drop(columns = ['Results', 
                                                     'Facility_Type', 
                                                     'LICENSE_DESCRIPTION', 
                                                     'WARD_PRECINCT'])


# get dummies
X = pd.get_dummies(X)
```

## 5. Discussion Questions

1. Why do we need metrics beyond accuracy when using machine learning in the social sciences and public policy?


Accuracy only tells us the percentage of correct predictions, but it completely ignores the costs of different types of errors, which can have vastly different implications in result fields. For instance, in our example of public health inspections, predicting that a failing restaurant will pass (false negative) is far worse than incorrectly flagging a passing restaurant for an audit (false positive). Metrics like precision and recall help unpack these nuances (i.e. precision ensures we don’t waste resources auditing businesses that don’t need it, recall ensures we catch as many problematic establishments as possible). It helps us figure out the direcitonality of our errors and the risks taken through using such a model.
